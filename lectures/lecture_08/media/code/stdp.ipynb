```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def W(θ, αp=1.0, αm=1.0, τp=10e-3, τm=10e-3):
    θ = np.atleast_1d(θ)
    res = (θ >  0.0) * αp * np.exp(-θ / τp) \
        - (θ <  0.0) * αm * np.exp( θ / τm)
    res[np.abs(θ) < 1e-6] = np.nan
    return res
```

```python
θs = np.linspace(-50e-3, 50e-3, 111)
fig, ax = plt.subplots(figsize=(6, 2.5))
ax.axhline(0, color='k', linestyle='--', linewidth=1.0)
ax.axvline(0, color='k', linestyle='--', linewidth=1.0)
ax.plot(θs * 1e3, W(θs), linewidth=3.0)
ax.set_xlabel("$\\theta = t^\mathrm{post} - t^\mathrm{pre}$ (ms)")
ax.set_ylabel("Synaptic weight change $\\Delta w_{ij}$")
ax.text(-55, -1.46, "$\\leftarrow$ Post before pre-spike", fontsize=8, ha="left", va="bottom")
ax.text(55, -1.46, "Post after pre-spike $\\rightarrow$", fontsize=8, ha="right", va="bottom")

fig.savefig('../stdp.pdf', bbox_inches='tight', transparent=True)
```
